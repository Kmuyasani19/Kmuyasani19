```python
# Cell 4: Complete FastAPI backend (UPDATED VERSION)
from fastapi import FastAPI, HTTPException
from fastapi.middleware.cors import CORSMiddleware
from pydantic import BaseModel
from typing import Optional, List, Dict, Any
import nest_asyncio
import uvicorn
import threading
import requests
import json

# FIXED IMPORTS - Using the correct paths for latest LangChain
# Instead of langchain.agents, use langchain_core or langchain.agents
try:
    # Try the newer import structure first
    from langchain_ollama import OllamaLLM
    from langchain_core.agents import AgentExecutor
    from langchain_core.tools import BaseTool
    from langchain_core.prompts import PromptTemplate
    LANGCHAIN_AVAILABLE = True
    print("✅ Using langchain_core imports")
except ImportError:
    # Fallback to older structure
    try:
        from langchain_community.llms import Ollama
        from langchain.agents import AgentExecutor, create_react_agent
        from langchain.prompts import PromptTemplate
        LANGCHAIN_AVAILABLE = True
        print("✅ Using langchain_community imports")
    except ImportError as e:
        print(f"⚠️ LangChain imports failed: {e}")
        LANGCHAIN_AVAILABLE = False

# For the simplest approach, we'll use direct Ollama calls without LangChain
# This is more reliable and avoids import issues

# Apply nest_asyncio to allow running event loop in Colab
nest_asyncio.apply()

# Create FastAPI app
app = FastAPI(title="MU Co-Pilot API", version="1.0.0")

# Add CORS middleware (allows your Flutter frontend to connect)
app.add_middleware(
    CORSMiddleware,
    allow_origins=["*"],  # In production, replace with your actual frontend URL
    allow_credentials=True,
    allow_methods=["*"],
    allow_headers=["*"],
)

# Request/Response models
class ChatRequest(BaseModel):
    message: str
    session_id: Optional[str] = None
    use_context: Optional[bool] = True

class ChatResponse(BaseModel):
    response: str
    session_id: str
    sources: List[str] = []

# Initialize Ollama connection
ollama_url = "http://localhost:11434"

def check_ollama():
    """Check if Ollama is running"""
    try:
        response = requests.get(f"{ollama_url}/api/tags", timeout=5)
        return response.status_code == 200
    except:
        return False

def generate_response(prompt: str, context: Optional[str] = None) -> str:
    """Generate response using Ollama directly (no LangChain dependencies)"""

    # Prepare the prompt with context if provided
    if context:
        full_prompt = f"""You are MU Co-Pilot, an AI assistant for Mulungushi University students.

Context information:
{context}

Student question: {prompt}

Please provide a helpful, accurate answer based on the context above. If the context doesn't contain the answer, say so politely and offer to help with other questions.

Answer:"""
    else:
        full_prompt = f"""You are MU Co-Pilot, an AI assistant for Mulungushi University students.

Student question: {prompt}

Please provide a helpful, accurate answer. If you don't know the answer, say so politely.

Answer:"""

    try:
        response = requests.post(
            f"{ollama_url}/api/generate",
            json={
                "model": "gemma:7b",  # or "mistral"
                "prompt": full_prompt,
                "stream": False,
                "options": {
                    "temperature": 0.3,
                    "num_predict": 512,
                    "top_p": 0.85,
                    "repeat_penalty": 1.1
                }
            },
            timeout=60
        )

        if response.status_code == 200:
            result = response.json()
            return result.get("response", "I apologize, but I couldn't generate a response.")
        else:
            return f"Error: Ollama returned status {response.status_code}"

    except requests.exceptions.ConnectionError:
        return "Error: Cannot connect to Ollama. Please make sure the service is running."
    except requests.exceptions.Timeout:
        return "Error: The request timed out. Please try again."
    except Exception as e:
        return f"Error: {str(e)}"

# Simple RAG with document context
class SimpleRAG:
    def __init__(self):
        self.knowledge_base = []

    def add_document(self, title: str, content: str):
        """Add a document to the knowledge base"""
        self.knowledge_base.append({
            "title": title,
            "content": content
        })

    def search(self, query: str, top_k: int = 2) -> List[str]:
        """Simple keyword-based search (you can upgrade to embeddings later)"""
        query_lower = query.lower()
        relevant_docs = []

        for doc in self.knowledge_base:
            if any(keyword in doc["content"].lower() for keyword in query_lower.split()):
                relevant_docs.append(doc["content"])

        return relevant_docs[:top_k]

# Initialize RAG with sample knowledge
rag = SimpleRAG()

# Add sample knowledge about Mulungushi University
sample_knowledge = """
Mulungushi University Admission Requirements:
- Five O-level credits including Mathematics and English
- Two A-level passes in relevant subjects
- Application fee: ZMW 200
- Deadline: 30th November each year
- Required documents: Certified copies of certificates, National Registration Card, passport photos

Computer Science Program Requirements:
- Mathematics O-level (Credit or better)
- Science subject at A-level (Physics, Chemistry, or Biology)
- English Language O-level (Credit or better)
- No prior programming experience required

Fee Structure:
- Zambian Students: ZMW 8,000 per semester
- International Students: USD 1,500 per semester
- Registration Fee: ZMW 500 (new students)
- Payment deadlines: 30th September (Semester 1), 31st January (Semester 2)

Academic Calendar:
- First Semester: September to December
- Second Semester: January to April
- Long Vacation: May to August
"""

rag.add_document("university_info", sample_knowledge)

# API Endpoints
@app.get("/")
async def root():
    return {
        "message": "MU Co-Pilot API is running",
        "status": "healthy",
        "ollama": check_ollama(),
        "model": "gemma:7b"
    }

@app.get("/health")
async def health():
    return {
        "status": "healthy",
        "ollama": check_ollama(),
        "model": "gemma:7b",
        "rag_documents": len(rag.knowledge_base)
    }

@app.post("/api/chat", response_model=ChatResponse)
async def chat(request: ChatRequest):
    """Main chat endpoint with RAG context"""
    if not check_ollama():
        raise HTTPException(status_code=503, detail="Ollama service not available")

    # Search for relevant context
    context_docs = rag.search(request.message)
    context = "\n\n".join(context_docs) if context_docs else None

    # Generate response with context
    response_text = generate_response(request.message, context)

    return ChatResponse(
        response=response_text,
        session_id=request.session_id or "default",
        sources=context_docs
    )

@app.post("/api/simple-chat")
async def simple_chat(request: ChatRequest):
    """Simple chat endpoint without RAG for testing"""
    if not check_ollama():
        return {"response": "Ollama is not ready yet. Please wait."}

    response_text = generate_response(request.message)
    return {"response": response_text}

@app.get("/api/status")
async def status():
    return {
        "status": "online",
        "model": "gemma:7b",
        "ollama_ready": check_ollama(),
        "rag_ready": len(rag.knowledge_base) > 0
    }

@app.post("/api/add-knowledge")
async def add_knowledge(title: str, content: str):
    """Add knowledge to the RAG system"""
    rag.add_document(title, content)
    return {"message": f"Added knowledge: {title}", "total_docs": len(rag.knowledge_base)}

print("✅ FastAPI app created successfully!")
```

```python
# Start the FastAPI server
import threading
import uvicorn

def run_server():
    uvicorn.run(app, host="0.0.0.0", port=8000, log_level="info")

# Start server in background thread
server_thread = threading.Thread(target=run_server, daemon=True)
server_thread.start()

print("✅ FastAPI server is running!")
print(f"📍 Public URL: {public_url}")
print(f"📚 API Docs: {public_url}/docs")
print(f"💚 Health Check: {public_url}/health")
print(f"💬 Chat endpoint: {public_url}/api/chat")
```

```python
# Keep session alive by pinging every 5 minutes
import time
import threading

def keep_alive():
    while True:
        time.sleep(300)  # 5 minutes
        try:
            requests.get(f"{public_url}/health")
            print("💓 Keep-alive ping sent")
        except:
            pass

alive_thread = threading.Thread(target=keep_alive, daemon=True)
alive_thread.start()
```
